```python
#Tool

from langchain_community.tools import ArxivQueryRun,WikipediaQueryRun
from langchain_community.utilities import ArxivAPIWrapper,WikipediaAPIWrapper
```

```python
api_wrapper_arxiv = ArxivAPIWrapper(top_k_results=2,doc_content_chars_max=500)
arxiv=ArxivQueryRun(api_wrapper=api_wrapper_arxiv,description="Query Arxiv for research papers")
tools=[arxiv]

api_wrapper_wikipedia = WikipediaAPIWrapper(top_k_results=2,doc_content_chars_max=500)
wikipedia=WikipediaQueryRun(api_wrapper=api_wrapper_wikipedia,description="Query Wikipedia for information")
tools=[wikipedia]
```

```python
arxiv.invoke("Explain quantum computing in simple terms")
```

```python
wikipedia.invoke("Explain pydantic in simple terms")
```
